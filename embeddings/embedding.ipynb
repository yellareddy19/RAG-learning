### This is creating embeddings for a sentance

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings=HuggingFaceEmbeddings(
    model="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
sample_text="Hi iam learning how to create embeddings. you know iam good at AI "

embedding=embeddings.embed_query(sample_text)
print(f"embeddings are {embedding}")
print(f"embeddings created of length {len(embedding)}")
```

```python
# for multiple sentances

sentance=["the tree is green in color",
          "the giry is in white color"
          "sky is dark"
          ]
embedding=embeddings.embed_documents(sentance)
print(f'embeddings are {embedding}')
print(f'length of each sentance embedding is {len(embedding[0])}')
```
