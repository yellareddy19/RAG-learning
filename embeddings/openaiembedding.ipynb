### Creating embeddings using openai

```python
### to use openai api key there are 2 ways one way is using directly mentioning api key 
# and other way is by loding through environment variable

#Loading api key through environment variable

import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings

embedding=OpenAIEmbeddings(
    model="text-embedding-3-small"
)
```

```python
text="This is openai embedding model"
embedding.embed_query(text)
embedding
```
