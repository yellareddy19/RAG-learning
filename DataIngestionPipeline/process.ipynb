```python
from  langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from typing import List
from langchain_core.documents import Document
class PdfParser:
    def __init__(self,chunk_size=100,chunk_overlap=10):
        self.chunk_size=chunk_size,
        self.chunk_overlap=chunk_overlap,
        self.text_splitter=RecursiveCharacterTextSplitter(
            chunk_size=chunk_size,
            chunk_overlap=chunk_overlap,
            separators=[" ",".","/n"]
            
        )

    def pdfLoader(self,filepath:str)->List[Document]:
        loader=PyPDFLoader(filepath)
        content=loader.load()

        processed_chunks=[]


        for counter,key in enumerate(content):
            CleanedText= self._CleanText(key.page_content)
        
            if len(CleanedText) < 100:
                continue

            # Creating chunks with enhancing metadata

            chunks=self.text_splitter.create_documents(
                texts=[CleanedText],
                metadatas=[{
                    **key.metadata,
                    "page": counter+1,
                    "total_pages":len(content),
                    "chunk_method": "PdfParser",
                    "char_count": len(CleanedText)




                }]

            )

            processed_chunks.extend(chunks)

        
        return processed_chunks

    def _CleanText(self,texts):
        text=" ".join(texts.split())
        return text


processor=PdfParser()




            


```

```python
smart_chunks=processor.pdfLoader("data/pdfs/Yellareddy.pdf")
print(type(smart_chunks))
print(len(smart_chunks))
print(smart_chunks)

##to see metadawhat we enhanced
print("-----------------------printing metadata-------------------------")
for key,value in smart_chunks[0].metadata.items():
    
    print(f"{key}:{value}")


for i in smart_chunks:
    print(f"{i.page_content}   :    {i.metadata}")
    print()

    
```
